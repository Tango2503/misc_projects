```python
def bar_plot(data,column_name):
    grouped = data.groupby([column_name])['Name'].count().reset_index()
    plt.bar(grouped[column_name], grouped['Name'], alpha = 0.9)
    # Get current axis on current figure
    ax = plt.gca()
    # Iterate through the list of axes' patches
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), fontsize=12, color='red', ha='center', va='bottom')
```

```python
data.columns
```

```python
bar_columns = ['Type_1', 'Generation', 'isLegendary','Color', 'hasGender', 'Egg_Group_1','hasMegaEvolution','Body_Style']
plt.figure(figsize = (15,35))
for i in range(1,len(bar_columns)+1):
    plt.subplot(len(bar_columns),1,i)
    plt.title(bar_columns[i-1])
    bar_plot(data,bar_columns[i-1])
plt.show()
    
```

```python
def hist_plot(data,column_name):
    data = data.dropna()
    #grouped = data.groupby([column_name])['Name'].count().reset_index()
    #plt.hist(grouped[column_name])
    sns.distplot(data[column_name], axlabel = False)
    

hist_columns = ['Total', 'HP', 'Attack','Defense', 'Sp_Atk', 'Sp_Def', 'Speed', 'Pr_Male', 'Height_m', 'Weight_kg', 'Catch_Rate']
plt.figure(figsize = (15,35))
for i in range(1,len(hist_columns)+1):
    plt.subplot(len(hist_columns),1,i)
    plt.title(hist_columns[i-1])
    hist_plot(data,hist_columns[i-1])
plt.show()
```

```python
def box_plot(data, column_name):
    grouped = data.groupby([column_name])['Name'].count().reset_index()
    fig = sns.boxplot(grouped['Name'])
    fig.set(title = column_name)
    fig.set(xlabel = None)
    

plt.figure(figsize = (15,35))
for i in range(1,len(hist_columns)+1):
    plt.subplot(len(hist_columns),1,i)
    plt.title(hist_columns[i-1])
    box_plot(data,hist_columns[i-1])
plt.show()
```

```python
from scipy import stats
from IPython.core import display as ICD

for i in hist_columns:
    outliers = data.loc[np.abs(stats.zscore(data[i]) > 3)]
    print(i,len(outliers))
    if outliers.size > 0:
        ICD.display(outliers)
    
```
